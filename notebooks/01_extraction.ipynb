# 01 — Extraction
Load the raw Olist tables, inspect structure, and confirm data is suitable for the project.

```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = (
    Path.cwd().resolve().parent
    if Path.cwd().resolve().name == 'notebooks'
    else Path.cwd().resolve()
)
RAW = PROJECT_ROOT / 'data' / 'raw'
```

```python
tables = {
    'orders'      : RAW / 'olist_orders_dataset.csv',
    'items'       : RAW / 'olist_order_items_dataset.csv',
    'customers'   : RAW / 'olist_customers_dataset.csv',
    'reviews'     : RAW / 'olist_order_reviews_dataset.csv',
    'products'    : RAW / 'olist_products_dataset.csv',
    'sellers'     : RAW / 'olist_sellers_dataset.csv',
    'translation' : RAW / 'product_category_name_translation.csv',
    'geolocation' : RAW / 'olist_geolocation_dataset.csv',
}

dfs = {name: pd.read_csv(path) for name, path in tables.items()}

for name, df in dfs.items():
    print(f'{name:15s}: {df.shape}')
```

```python
for name, df in dfs.items():
    nulls = df.isnull().sum().sum()
    dups  = df.duplicated().sum()
    print(f'{name:15s}: nulls={nulls:,}  duplicates={dups:,}')
```
